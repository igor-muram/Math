```python
from mpl_toolkits import mplot3d
%matplotlib qt5
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

# массив вершин призмы
v = np.array([[30, 30, 20], [30, 80, 20], [70, 80, 20],  [70, 30, 20], [30, 30, 170], [30, 80, 170], [70, 80, 110],  [70, 30, 110]])

# массив граней призмы
verts = [[v[0],v[1],v[5],v[4]], [v[0],v[3],v[7],v[4]], [v[2],v[3],v[7],v[6]], [v[1],v[2],v[6],v[5]], [v[0],v[1],v[2],v[3]], [v[4],v[5],v[6],v[7]]]

# массив 
shadow = [[v[x] - [0, 0, 15] for x in range(4)]]

fig = plt.figure()

# полноэкранный режим
plt.get_current_fig_manager().window.showMaximized() 

# призма
ax = fig.add_subplot(111, projection = '3d') 
ax.scatter3D(v[:, 0], v[:, 1], v[:, 2])

# параметры призмы
collection = Poly3DCollection(verts, linewidths = 3, alpha = .25)
collection.set_facecolor('cyan')
collection.set_edgecolor('black')

# добавление призмы и тени
ax.add_collection3d(collection)
ax.add_collection3d(Poly3DCollection(shadow, facecolor = 'grey', linewidths=1, alpha=.2))

# установка размеров осей
ax.set_zlim(0, 170)
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)

# подписи осей
ax.set_xlabel('x')
ax.set_ylabel('y')

# класс для создания стрелок
class Arrow3D(FancyArrowPatch):

    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0, 0), (0, 0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        FancyArrowPatch.draw(self, renderer)
    
# создание сетки для разбиения
Xpoints = [x * 10 for x in range(3, 8)] 
Ypoints = [y * 10 for y in range(3, 9)]
xv, yv = np.meshgrid(Xpoints, Ypoints)
ax.scatter3D(xv[:, :], yv[:, :], 20)

ycoord = 30
for x in xv:
    for y in x:
        z = 215 - 1.5*y
        
        # рисование стрелок
        a = Arrow3D([y, y], [ycoord, ycoord], [20, z], mutation_scale = 20, lw = 2, arrowstyle = "->", color = "k")
        ax.add_artist(a)
        ax.scatter3D(y, ycoord, z)
        
    ycoord = ycoord + 10
```
